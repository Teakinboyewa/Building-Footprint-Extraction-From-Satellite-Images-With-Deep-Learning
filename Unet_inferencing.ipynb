# Importing Mendatory modules

```python
from PIL import Image
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from numpy import asarray
```

# Loading Models

```python
model = keras.models.load_model('/content/unet_building_detection_epoch1000.h5')
```

Upload your inference image

```python
infer_img = Image.open('/content/fmi/processed_image+final_00012_a0.png')
plt.figure(figsize=(10,10))
plt.imshow(infer_img)
img = asarray(infer_img)
img = np.expand_dims(img,axis=0)
```

Predictions

```python
pred = model.predict(img,verbose=1)
preds = np.argmax(pred[0],axis=-1)
plt.figure(figsize=(10,10))
plt.imshow(preds)
```

# Plotting and Visualization

```python
plt.figure(figsize=(15,15))
plt.subplot(121)
plt.imshow(preds)
plt.title("Predictions on Infereced Image")
plt.subplot(122)
plt.imshow(np.squeeze(img))
plt.title("Inferenced RGB image")
```
